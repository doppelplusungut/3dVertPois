```python
import pandas as pd
```

```python
poi_df = pd.read_csv('/home/daniel/Documents/Uni/MT/poi-prediction/data/dir-las_scale-iso/poi_distances.csv')
```

```python
#Remove all rows with NaN values in the gruber_surface_distance column
poi_df = poi_df.dropna(subset=['gruber_surface_distance'])

#Only keep the following columns: subject_name, vertebra, poi_type, gruber_x, gruber_y, gruber_z, gruber_surface_distance, gruber_centroid_distance
poi_df = poi_df[['subject_name', 'vertebra', 'poi_type', 'gruber_x', 'gruber_y', 'gruber_z', 'gruber_surface_distance', 'gruber_centroid_distance']]

poi_df.head()
```

```python
#Get the gruber_surface_distance mean and std overall
overall_stats = poi_df['gruber_surface_distance'].describe()

#Get the gruber_surface_distance mean, std, min, max, median, 5th, 25th, 75th, 95th percentile for each subject in poi_df
per_sub_stats = poi_df.groupby('subject_name')['gruber_surface_distance'].describe(percentiles = [.01, .05, .25, .75, .95, .99])

#Get the gruber_surface_distance mean, std, min, max, median, 5th, 25th, 75th, 95th percentile for each vertebra in poi_df
per_vert_stats = poi_df.groupby('vertebra')['gruber_surface_distance'].describe(percentiles = [.01, .05, .25, .75, .95, .99])

#Get the gruber_surface_distance mean, std, min, max, median, 5th, 25th, 75th, 95th percentile for each poi_type in poi_df
per_poi_stats = poi_df.groupby('poi_type')['gruber_surface_distance'].describe(percentiles = [.01, .05, .25, .75, .95, .99])
```

```python
#Sort the per_sub_stats by the mean of the gruber_surface_distance
per_sub_stats = per_sub_stats.sort_values(by=['mean'], ascending=False)

per_sub_stats
```

```python
#Remove WS-09, WS-05, WS-17, WS-07 and WS-08 from the dataset since they have a significantly higher mean gruber_surface_distance than the other subjects
bad_subjects = ['WS-09', 'WS-05', 'WS-17', 'WS-07', 'WS-08']
poi_df = poi_df[~poi_df.subject_name.isin(bad_subjects)]
```

```python
#Calculate per_poi_stats and per_vert_stats again
per_vert_stats = poi_df.groupby('vertebra')['gruber_surface_distance'].describe(percentiles = [.01, .05, .25, .75, .95, .99])
per_poi_stats = poi_df.groupby('poi_type')['gruber_surface_distance'].describe(percentiles = [.01, .05, .25, .75, .95, .99])
```

```python
per_poi_stats = per_poi_stats.sort_values(by=['mean'], ascending=False)

per_poi_stats
```

```python
per_vert_stats = per_vert_stats.sort_values(by=['mean'], ascending=False)

per_vert_stats
```

Vertebrae 17, 20 and 23 only have bad averages because of crazy high outliers, but overall the distribution seems somewhat okay. Vertebra 7 has a very high 95-percentile. It might be worth digging into that.

```python
#Group the poi_df by subject_name and vertebra and get the number of rows in each group as well as the number of groups
per_sub_vert = poi_df.groupby(['subject_name', 'vertebra']).size().reset_index(name='counts')

print('Number of subject-vertebra pairs: ' + str(len(per_sub_vert)))

#Get the number of subject-vertebra pairs with a count other than 23
print('Number of subject-vertebra pairs with a count other than 23: ' + str(len(per_sub_vert[per_sub_vert['counts'] != 23])))
```

```python
#Get a list of subjects with a mean gruber_surface_distance of <2 mm and std <2 mm
good_subjects = per_sub_stats[(per_sub_stats['mean'] < 2) & (per_sub_stats['std'] < 2)].index.tolist()

#Select 5 random elements from good_subjects
import random
random.seed(42)
test_subjects = random.sample(good_subjects, 5)

test_subjects
```

```python
#From the remaining subjects, select 5 random elements
remaining_subjects = list(set(per_sub_stats.index.tolist()) - set(test_subjects))
remaining_subjects = list(set(remaining_subjects) - set(bad_subjects))

random.seed(42)
val_subjects = random.sample(remaining_subjects, 5)

val_subjects
```

```python
#Add a column split to poi_df. The value is "test" if subject is in test_subjects, "val" if subject is in val_subjects, and "train" otherwise
poi_df['split'] = 'train'
poi_df.loc[poi_df['subject_name'].isin(test_subjects), 'split'] = 'test'
poi_df.loc[poi_df['subject_name'].isin(val_subjects), 'split'] = 'val'
```

```python
#Count the number of train, val, and test samples
poi_df['split'].value_counts()
```

```python
#Get a list of all subject-vertebra pairs where at least one poi has a gruber_surface distance of >4 mm
bad_pairs = poi_df[poi_df['gruber_surface_distance'] > 3][['subject_name', 'vertebra']].drop_duplicates().values.tolist()
len(bad_pairs)
```

```python
#This might be too aggressive since we drop almost 50% of the data, but we well try it for now

#Create a new column called bad_pair in poi_df. If the subject_name and vertebra are in bad_pairs, the value is True, otherwise False. Also create a column
#called comment which contains 'gruber_surface_distance > 5 mm' for all rows where bad_pair is True

poi_df['good_pair'] = poi_df.apply(lambda x: False if [x['subject_name'], x['vertebra']] in bad_pairs else True, axis=1)
poi_df['comment'] = poi_df.apply(lambda x: 'gruber_surface_distance > 5 mm for at least one poi on this sub-vert pair' if not x['good_pair'] else '', axis=1)

poi_df.head()
```

```python
#Count the number of rows where bad_pair is True and where bad_pair is False
poi_df['good_pair'].value_counts()
```
